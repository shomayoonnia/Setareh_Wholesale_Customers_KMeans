```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist, pdist

# Load dataset
df = pd.read_csv("Wholesale customers data.csv")

# Drop missing values
df = df.dropna()

# Feature selection
X = df.drop(columns=["Channel","Region"])

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dunn Index function
def dunn_index(X, labels):

    unique_clusters = np.unique(labels)
    clusters = [X[labels == k] for k in unique_clusters]

    # inter cluster distance
    intercluster = np.min([
        cdist(c1, c2).min()
        for i, c1 in enumerate(clusters)
        for c2 in clusters[i+1:]
    ])

    # intra cluster distance
    intracluster = np.max([
        pdist(c).max()
        for c in clusters
        if len(c) > 1
    ])

    return intercluster / intracluster


# Run KMeans for different k
K = range(2,11)

inertia_values = []
dunn_values = []

for k in K:

    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)

    inertia = kmeans.inertia_
    inertia_values.append(inertia)

    dunn = dunn_index(X_scaled, labels)
    dunn_values.append(dunn)

    print(f"k = {k} | inertia = {inertia:.2f} | dunn index = {dunn:.4f}")

# Elbow curve
plt.figure(figsize=(6,4))
plt.plot(K, inertia_values, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Curve")
plt.show()

# choose best k from elbow
best_k = 3
print("Best k chosen:", best_k)

# Final model
model = KMeans(n_clusters=best_k, random_state=42)
clusters = model.fit_predict(X_scaled)

# Predict clusters for first 10 samples
print("Predicted clusters for first 10 samples:")
print(clusters[:10])
```
